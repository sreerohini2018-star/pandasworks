```python
#platformwise post count
df["platform"].value_counts()
```

```python
#filter post with more than 1m(1_000_000)views
df[df["views"]>1000000]
```

```python
#average engagement rate by platform
df.groupby("platform")["engagement_rate"].mean()
```

```python
#top3 post with most likes
df.sort_values(by="likes",ascending=False).head(3)
```

```python
#total views by region
df.groupby("region")["views"].sum()
```

```python
#which content type has highest engagement rate
df.groupby("content_type")["engagement_rate"].sum().idxmax()
```

```python
#count of viral posts
df["is_viral"].value_counts()
```

```python
#find average sentiment score of viralpost
df[df["is_viral"]==1]["sentiment_score"].mean()
```

```python
#total interaction
df["total_interaction"]=df["likes"]+df["shares"]+df["comments"]
df.head(3)
```

```python
#adding new coloumn
df["post_datetime"]=pd.to_datetime(df["post_datetime"])
df["month"]=df["post_datetime"].dt.month
df.head(3)
```

```python
#number of post in each month
df["month"].value_counts()
```

```python
#which factor makes a post viral
df.groupby("is_viral")[["likes","comments","shares","engagement_rate"]].mean()
```

```python
#which region produces more viral post
df.groupby("region")["is_viral"].sum().idxmax()
#df[df["is_viral"]==1]["region"].value_counts().head(1)
```
