```python
import pandas as pd
df=pd.read_csv("C:\\Users\\sreer\\Desktop\\MLJOURNEY\\pandas_works\\world_famous_places\\world_famous_places_2024.csv")
df.head(3)
```

```python
#which attraction has the highest annual visitors?
df.loc[df["Annual_Visitors_Millions"].idxmax(), ["Place_Name", "Annual_Visitors_Millions"]]
```

```python
#list all attractions in France
df[df["Country"] == "France"][["Place_Name", "City"]]
```

```python
#how many UNESCO World Heritage sites are there?
df["UNESCO_World_Heritage"].value_counts()
```

```python
#average annual visitors by region
df.groupby("Region")["Annual_Visitors_Millions"].mean().sort_values(ascending=False)
```

```python
#top 5 attractions by tourism revenue
df.sort_values("Tourism_Revenue_Million_USD", ascending=False).head(5)
```

```python
#attractions with free entry
df[df["Entry_Fee_USD"] == 0][["Place_Name", "Country", "Annual_Visitors_Millions"]]
```

```python
#city with the most attractions
df["City"].value_counts().head(1)
```

```python
#average visit duration by attraction type
df.groupby("Type")["Average_Visit_Duration_Hours"].mean().sort_values(ascending=False)
```

```python
#UNESCO vs Non-UNESCO: average visitors
df.groupby("UNESCO_World_Heritage")["Annual_Visitors_Millions"].mean()
```

```python
#High visitors (>10M) but low visit duration (<2 hours)
df[
    (df["Annual_Visitors_Millions"] > 10) &
    (df["Average_Visit_Duration_Hours"] < 2)
][["Place_Name", "Annual_Visitors_Millions", "Average_Visit_Duration_Hours"]]
```
